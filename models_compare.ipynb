```python
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
```

```python
testing_data_gen= ImageDataGenerator(rescale = 1./255, 
                                   )

model1_testing = testing_data_gen.flow_from_directory('datasetforproject/testing_dataset/women_model',
                                                 target_size = (48,48), 
                                                 color_mode = 'grayscale' , 
                                                 class_mode = 'categorical')
model2_testing = testing_data_gen.flow_from_directory('datasetforproject/testing_dataset/men_model',
                                                 target_size = (48,48), 
                                                 color_mode = 'grayscale' , 
                                                 class_mode = 'categorical')
model_filenames = ["my_model1.h5", "my_model1_aug.h5", "my_model1_dropout.h5","my_model1_dropout_aug.h5","my_model1_dropout_lr.h5",
                   "model1_cnn_dropout_rg.h5","model1_cnn_dropout_rg_v2.h5"]

# Lists to store results
losses_women = []
accuracies_women = []


for model_filename in model_filenames:
    # Load the model
    model = load_model(model_filename)

    # Evaluate the model
    loss, accuracy = model.evaluate(model1_testing, verbose=2)

    # Store the results
    losses_women.append(loss)
    accuracies_women.append(accuracy)

    # Print or store the evaluation results
    print(f"Model {model_filename} - Loss: {loss}, Accuracy: {accuracy}")


```

```python
# Plotting
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(30, 12))

# Plot Losses
ax1.bar(model_filenames, losses_women, color='blue')
ax1.set_title('Losses')
ax1.set_ylabel('Loss')

# Plot Accuracies
ax2.bar(model_filenames, accuracies_women, color='green')
ax2.set_title('Accuracies')
ax2.set_ylabel('Accuracy')

plt.show()
```
